```python

class Vector3(object):
  def __init__(self, x = 0, y = 0, z = 0):
    self._data = np.array([x,y,z], dtype = "float32")

  @property
  def x(self):
      """Property x is first element of vector.

      >>> Vector3(10, 20, 30).x
      10.0
      """
      return self._data[0]

  @x.setter
  def x(self, value):
    self._data[0] = value

  @property
  def y(self):
      """Property y is first element of vector.

      >>> Vector3(10, 20, 30).y
      20.0
      """
      return self._data[1]

  @y.setter
  def y(self, value):
    self._data[1] = value

  @property
  def z(self):
      """Property z is first element of vector.

      >>> Vector3(10, 20, 30).z
      30.0
      """
      return self._data[2]

  @z.setter
  def z(self, value):
    self._data[2] = value

  def _np_wrapper(self,npfunc,value1, value2):
    if isinstance(value1, type(self)):
      value1 = value1._data
    if isinstance(value2, type(self)):
      value2 = value2._data

    result = npfunc(value1,value2)
    return type(self)(x = result[0], y = result[1], z = result[2])

  def __str__(self):
    return str('[{0:.4f}, {1:.4f}, {2:.4f}]'.format(*self._data))

  def __repr__(self):
    return 'Vector3({0}, {1}, {2})'.format(*self._data)

  def __mul__(self, value):
    return self._np_wrapper(np.multiply,self._data,value)

  def __rmul__(self, value):
    return self.__mul__(value) #Commutative

  def __add__(self, value):
    return self._np_wrapper(np.add,self._data,value)

  def __radd__(self, value):
    return self.__add__(value) #Commutative

  def __sub__(self, value):
      return self._np_wrapper(np.subtract,self._data,value)

  def __rsub__(self, value):
    return self._np_wrapper(np.subtract,value,self._data)

  def Dot(self, value):
    if isinstance(value, type(self)):
      value = value._data

    return np.dot(self._data,value)

  def Cross(self, value):
    return self._np_wrapper(np.cross,self._data,value)

  def Magnitude(self):
    return np.sqrt(self.Dot(self._data))

  def Normalized(self):
    return self * (1/self.Magnitude())

  def ProjectToVector(self, value):
    if isinstance(value, type(self)):
      value = value._data

    result = value * self.Dot(value) / np.dot(value, value)

    return type(self)(x = result[0], y = result[1], z = result[2])

  def ProjectToPlane(self, normal):
    return self - self.ProjectToVector(normal)

  def Get2DOverTopPlane(self):
    return Vector2(self.x, self.z)

#Test Vector3
a = Vector3(1.09,2.10,3.21)
b = Vector3(4.32,5.43,6.54)
c = Vector3(0,0,0)
print(a+b)
print(b+a)
print(a*b)
print(b*a)
print(a-b)
print(b-a)
print(a.Dot(b))
print(b.Dot(a))
print(a.Cross(b))
print(b.Cross(a))
print(a.Magnitude())
print(b.Magnitude())
print(c.Magnitude())
print(a.Normalized())
print(b.Normalized())
print(c.Normalized())
print("projections")
print(a.ProjectToVector(b))
print(b.ProjectToVector(a))
print(a.ProjectToVector(c))
print(c.ProjectToVector(a))
print("pPlanes")
print(a.ProjectToPlane(b))
print(b.ProjectToPlane(a))
print(a.ProjectToPlane(c))
print(c.ProjectToPlane(a))
print(a.ProjectToPlane(Vector3(0,1,0)))
print(b.ProjectToPlane(Vector3(0,1,0)))
print(a.Get2DOverTopPlane())
print(b.Get2DOverTopPlane())





```

```python
from pyquaternion import Quaternion
import math as mt

#https://stackoverflow.com/questions/12088610/conversion-between-euler-quaternion-like-in-unity3d-engine

def EulerToQuaternion(v):
  xOver2 = mt.radians(v.x) * 0.5;
  yOver2 = mt.radians(v.y) * 0.5;
  zOver2 = mt.radians(v.z) * 0.5;

  sinXOver2 = mt.sin(xOver2);
  cosXOver2 = mt.cos(xOver2);
  sinYOver2 = mt.sin(yOver2);
  cosYOver2 = mt.cos(yOver2);
  sinZOver2 = mt.sin(zOver2);
  cosZOver2 = mt.cos(zOver2);

  w = cosYOver2 * cosXOver2 * cosZOver2 + sinYOver2 * sinXOver2 * sinZOver2
  x = cosYOver2 * sinXOver2 * cosZOver2 + sinYOver2 * cosXOver2 * sinZOver2
  y = sinYOver2 * cosXOver2 * cosZOver2 - cosYOver2 * sinXOver2 * sinZOver2
  z = cosYOver2 * cosXOver2 * sinZOver2 - sinYOver2 * sinXOver2 * cosZOver2

  return Quaternion([w,x,y,z])

def QuaternionToEuler(q):
  # if the input quaternion is normalized, this is exactly one. Otherwise, this acts as a correction factor for the quaternion's not-normalizedness
  unit = (q.x * q.x) + (q.y * q.y) + (q.z * q.z) + (q.w * q.w)
  # this will have a magnitude of 0.5 or greater if and only if this is a singularity case
  test = q.x * q.w - q.y * q.z
  if test > 0.4995 * unit: #singularity at north pole
    x = mt.pi / 2
    y = 2 * mt.atan2(q.y, q.x)
    z = 0
  elif test < -0.4995 * unit: # singularity at south pole
    x = -mt.pi/ 2
    y = -2 * mt.atan2(q.y, q.x)
    z = 0
  else:  #no singularity - this is the majority of cases
    x = mt.asin(2 * (q.w * q.x - q.y * q.z))
    y = mt.atan2(2 * q.w * q.y + 2 * q.z * q.x, 1 - 2 * (q.x * q.x + q.y * q.y))
    z = mt.atan2(2 * q.w * q.z + 2 * q.x * q.y, 1 - 2 * (q.z * q.z + q.x * q.x))

  return Vector3 (mt.degrees(x)%360, mt.degrees(y)%360, mt.degrees(z)%360)

#Test to compare with Unity Quaternions. Good enough.
for i in range(0,360,45):
    for j in range(0,360,45):
        for k in range(0,360,45):
          v1 = Vector3(i, j, k)
          q = EulerToQuaternion(v1)
          v2 = QuaternionToEuler(q)
          qp =  EulerToQuaternion(v2)
          print(str(v1) + " " + str(['%.1f' % n for n in q]) + " " + str(v2) + " " + str(['%.1f' % n for n in qp]) + "\n")

def GetDeltaRotationOverTopPlane(quat1,quat2):
  vec1 = quat1.rotate([0,0,1]) #ForwardVector
  vec2 = quat2.rotate([0,0,1]) #ForwardVector
  vector1 = Vector3(x = vec1[0], y = vec1[1], z = vec1[2])
  vector2 = Vector3(x = vec2[0], y = vec2[1], z = vec2[2])
  vector1 = vector1.Get2DOverTopPlane()
  vector2 = vector2.Get2DOverTopPlane()
  return mt.degrees(vector1.SignedAngle(vector2))

#Test
print("Angle between cuaternions top-view")
q1 = EulerToQuaternion(Vector3(0,0,0))
q2 = EulerToQuaternion(Vector3(0,90,0))

print(GetDeltaRotationOverTopPlane(q1,q2))
print(GetDeltaRotationOverTopPlane(q2,q1))
```

```python
# Features:
# Yaw: Valor de 0-360 (o -180 a 180) rotacion sobre un eje.     En short term y longterm se calculan diferente.
# Pitch: Valor de 0-360 (o -180 a 180) rotación sobre un eje.     En short term y longterm se calculan diferente.
# VelocityX: Velocidad sobre el eje X, expresada en espacio recorrido en un tiempo determinado.
# VelocityY: Velocidad sobre el eje Y.

# Labels
# No feature. Output.
# fX: Valor posicional calculado como delta de la posición pero siguiendo ciertos calculos de rotación.
# fY: Valor posicional caluclado

# Tras tratar el output, el error se mide con F mayuscula.
# FX: Delta de la posición no calculado con datos, sino con la red.
# FY: Delta de la posición no calculado con datos.

# Compute Long-term features
from tqdm import tqdm
import math

def meanVect(xArr,yArr): #Equation (3) in the paper.
    x_delta = 0
    y_delta = 0

    for idx in range(0,len(xArr)-1):
      x_delta = xArr[idx+1]-xArr[idx]
      y_delta = yArr[idx+1]-yArr[idx]

    x_delta /= len(xArr)-1
    y_delta /= len(xArr)-1

    return Vector2 (x_delta,y_delta)

def longTermDatasetDeltaPos(user_id, df, window = 50, look_ahead = 50):

    print(f'Processing user:\t{user_id}')

    user_data = df[df["UserID"] == user_id].reset_index(drop=True)
    user_data = user_data[["Time_since_startup",
                           "Oculus Quest_DeviceRotation_w", "Oculus Quest_DeviceRotation_x", "Oculus Quest_DeviceRotation_y", "Oculus Quest_DeviceRotation_z",
                           "Oculus Quest_DevicePosition_x", "Oculus Quest_DevicePosition_y","Oculus Quest_DevicePosition_z",
                           "Oculus Quest_Forward_x","Oculus Quest_Forward_y","Oculus Quest_Forward_z"]]
    #user_data = user_data.applymap(clean_to_numeric)
    print ("User data registers: " + str(len(user_data)))

    user_data_quat = user_data.apply(lambda r: QuaternionToEuler(Quaternion([r["Oculus Quest_DeviceRotation_w"], r["Oculus Quest_DeviceRotation_x"], r["Oculus Quest_DeviceRotation_y"], r["Oculus Quest_DeviceRotation_z"]])), axis=1)

    pitchAcc = user_data_quat.apply(lambda r: math.radians(r.x))
    yawAcc = user_data_quat.apply(lambda r: math.radians(r.y))

    #-window: Because we need that data. -look_ahead: Because we cannot compare with non-existent data.
    LTIpitch = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIyaw = np.zeros(user_data.shape[0] - window - look_ahead)

    LTIForwardX = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIForwardY = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIForwardZ = np.zeros(user_data.shape[0] - window - look_ahead)

    LTIVelocityX = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIVelocityY = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIVelocityZ = np.zeros(user_data.shape[0] - window - look_ahead)

    LTIfX_prev = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIfY_prev = np.zeros(user_data.shape[0] - window - look_ahead)
    LTIfZ_prev = np.zeros(user_data.shape[0] - window - look_ahead)

    LTILabelfX = np.zeros(user_data.shape[0] - window - look_ahead)
    LTILabelfZ = np.zeros(user_data.shape[0] - window - look_ahead)

#Ryaw = velocity01.SignedAngle(Vector2(0,1)) #Simple Ryaw to calculate referenced velocity

    for i in tqdm(range(window, user_data.shape[0]-look_ahead), desc='Computing Long Term Features and Labels'):

        window_data = user_data[i-window:i]
        #i is current index, i+n what we want to calculate

        x_window = window_data["Oculus Quest_DevicePosition_x"].values #window start index SOULD BE INCLUDED i
        y_window = window_data["Oculus Quest_DevicePosition_z"].values

        vec_delta_mean = meanVect(x_window, y_window)

        Ryaw = vec_delta_mean.SignedAngle(Vector2(0,1))

        LTIyaw_current = yawAcc[i] - Ryaw
        LTIyaw[i-window] = LTIyaw_current
        LTIpitch[i-window] = pitchAcc[i]

        # velocity XY
        deltaTime12 = user_data["Time_since_startup"].astype(float).iloc[i] - user_data["Time_since_startup"].astype(float).iloc[i-1]
        position1 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i-1],user_data["Oculus Quest_DevicePosition_y"].iloc[i-1] ,user_data["Oculus Quest_DevicePosition_z"].iloc[i-1])
        position2 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i], user_data["Oculus Quest_DevicePosition_y"].iloc[i], user_data["Oculus Quest_DevicePosition_z"].iloc[i])
        F_prev = position2 - position1
        velocity12 = Vector3(F_prev.x / deltaTime12, F_prev.y / deltaTime12,F_prev.z / deltaTime12)

        # LTIvelocityX
        LTIVelocityX[i-window] = math.cos(-Ryaw) * velocity12.x - math.sin(-Ryaw) * velocity12.z
        LTIVelocityY[i-window] = velocity12.y
        LTIVelocityZ[i-window] = math.sin(-Ryaw) * velocity12.x + math.cos(-Ryaw) * velocity12.z

        # FORWARD VECTOR
        forward_current = Vector3 (user_data["Oculus Quest_Forward_x"].iloc[i], user_data["Oculus Quest_Forward_y"].iloc[i], user_data["Oculus Quest_Forward_z"].iloc[i])
        LTIForwardX[i-window] = math.cos(-Ryaw) * forward_current.x - math.sin(-Ryaw) * forward_current.z
        LTIForwardY[i-window] = forward_current.y
        LTIForwardZ[i-window] = math.sin(-Ryaw) * forward_current.x + math.cos(-Ryaw) * forward_current.z

        # LTIfX
        LTIfX_prev[i-window] = math.cos(-Ryaw) * F_prev.x - math.sin(-Ryaw) * F_prev.z
        LTIfY_prev[i-window] = F_prev.y
        LTIfZ_prev[i-window] = math.sin(-Ryaw) * F_prev.x + math.cos(-Ryaw) * F_prev.z

        # esto deben ser las etiquetas
        position3 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i+look_ahead], user_data["Oculus Quest_DevicePosition_y"].iloc[i+look_ahead],user_data["Oculus Quest_DevicePosition_z"].iloc[i+look_ahead])
        F = position3 - position2

        # LTIfX
        LTILabelfX[i-window] = math.cos(-Ryaw) * F.x - math.sin(-Ryaw) * F.z
        LTILabelfZ[i-window] = math.sin(-Ryaw) * F.x + math.cos(-Ryaw) * F.z

    featuresLT = {}
    featuresLT['LTIpitch'] = LTIpitch
    featuresLT['LTIyaw'] = LTIyaw

    featuresLT['LTIForwardX'] = LTIForwardX
    featuresLT['LTIForwardY'] = LTIForwardY
    featuresLT['LTIForwardZ'] = LTIForwardZ

    featuresLT['LTIfX_prev'] = LTIfX_prev
    featuresLT['LTIfY_prev'] = LTIfY_prev
    featuresLT['LTIfZ_prev'] = LTIfZ_prev

    featuresLT['LTIVelocityX'] = LTIVelocityX
    featuresLT['LTIVelocityY'] = LTIVelocityY
    featuresLT['LTIVelocityZ'] = LTIVelocityZ

    featuresLT['LTIlabelfX'] = LTILabelfX
    featuresLT['LTIlabelfZ'] = LTILabelfZ

#desnormalizar->Cambiar al sistema de referencia global-> Con esa posición, calcular la diferencia con el valor real.

    return pd.DataFrame(featuresLT)
```

## 3.5.2 Short-term (with a twist)

```python
# Features:
# Yaw: Valor de 0-360 (o -180 a 180) rotacion sobre un eje.     En short term y longterm se calculan diferente.
# Pitch: Valor de 0-360 (o -180 a 180) rotación sobre un eje.     En short term y longterm se calculan diferente.
# VelocityX: Velocidad sobre el eje X, expresada en espacio recorrido en un tiempo determinado.
# VelocityY: Velocidad sobre el eje Y.

# Labels
# No feature. Output.
# fX: Valor posicional calculado como delta de la posición pero siguiendo ciertos calculos de rotación.
# fY: Valor posicional caluclado

# Tras tratar el output, el error se mide con F mayuscula.
# FX: Delta de la posición no calculado con datos, sino con la red.
# FY: Delta de la posición no calculado con datos.

# Compute Short-term features
from tqdm import tqdm
import math

def shortTermDatasetDeltaPos(user_id,df):

    print(f'Processing user:\t{user_id}')

    user_data = df[df["UserID"] == user_id].reset_index(drop=True)
    user_data = user_data[["Time_since_startup",
                           "Oculus Quest_DeviceRotation_w", "Oculus Quest_DeviceRotation_x", "Oculus Quest_DeviceRotation_y", "Oculus Quest_DeviceRotation_z",
                           "Oculus Quest_DevicePosition_x", "Oculus Quest_DevicePosition_y","Oculus Quest_DevicePosition_z",
                           "Oculus Quest_Forward_x","Oculus Quest_Forward_y","Oculus Quest_Forward_z"]]

    user_data_quat = user_data.apply(lambda r: QuaternionToEuler(Quaternion([r["Oculus Quest_DeviceRotation_w"], r["Oculus Quest_DeviceRotation_x"], r["Oculus Quest_DeviceRotation_y"], r["Oculus Quest_DeviceRotation_z"]])), axis=1)

    pitchAcc = user_data_quat.apply(lambda r: math.radians(r.x))
    yawAcc = user_data_quat.apply(lambda r: math.radians(r.y))

    STIpitch = np.zeros(user_data.shape[0]-2-1) #2 vecause we need last frame velocity and the last-last frame veloctity.
    STIyaw = np.zeros(user_data.shape[0]-2-1)

    STIForwardX = np.zeros(user_data.shape[0] - 2 - 1)
    STIForwardY = np.zeros(user_data.shape[0] - 2 - 1)
    STIForwardZ = np.zeros(user_data.shape[0] - 2 - 1)

    STIvelocityX = np.zeros(user_data.shape[0]-2-1)
    STIvelocityY = np.zeros(user_data.shape[0]-2-1)
    STIvelocityZ = np.zeros(user_data.shape[0]-2-1)

    STIfX_prev = np.zeros(user_data.shape[0]-2-1)
    STIfY_prev = np.zeros(user_data.shape[0]-2-1)
    STIfZ_prev = np.zeros(user_data.shape[0]-2-1)

    STIlabelfX = np.zeros(user_data.shape[0]-2-1)
    STIlabelfZ = np.zeros(user_data.shape[0]-2-1)

    for i in tqdm(range(2,user_data.shape[0]-1), desc='Computing Sort Term Features and Labels'):

        #i is current index. i+n it what we want to calculate.

        # velocity XY
        deltaTime12 = user_data["Time_since_startup"].astype(float).iloc[i] - user_data["Time_since_startup"].astype(float).iloc[i-1]
        position1 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i-1], user_data["Oculus Quest_DevicePosition_y"].iloc[i-1],user_data["Oculus Quest_DevicePosition_z"].iloc[i-1])
        position2 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i], user_data["Oculus Quest_DevicePosition_y"].iloc[i],user_data["Oculus Quest_DevicePosition_z"].iloc[i])
        F_prev = position2 - position1
        velocity12 = Vector3(F_prev.x / deltaTime12, F_prev.y / deltaTime12, F_prev.z / deltaTime12)

        #new Ryaw:
        deltaTime01 = user_data["Time_since_startup"].astype(float).iloc[i-1] - user_data["Time_since_startup"].astype(float).iloc[i-1-1]
        position0 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i-1-1], user_data["Oculus Quest_DevicePosition_y"].iloc[i-1-1],user_data["Oculus Quest_DevicePosition_z"].iloc[i-1-1])
        velocity01 = Vector2((position1 - position0).x / deltaTime01, (position1 - position0).z / deltaTime01)
        Ryaw = velocity01.SignedAngle(Vector2(0,1)) #(Reciebed in RADIANS) Simple Ryaw to calculate referenced velocity

        #FEATURE: SHORT TERM INPUT PITCH AND YAW (HEAD)
        STIpitch[i-2] = pitchAcc[i] #euler.y
        STIyaw[i-2] = yawAcc[i] - Ryaw #euler.z - Ryaw

        #FEATURE: SHORT TERM INPUT velocity.
        STIvelocityX[i-2] = math.cos(-Ryaw) * velocity12.x - math.sin(-Ryaw) * velocity12.z
        STIvelocityY[i-2] = velocity12.y
        STIvelocityZ[i-2] = math.sin(-Ryaw) * velocity12.x + math.cos(-Ryaw) * velocity12.z

        # FORWARD VECTOR
        forward_current = Vector3 (user_data["Oculus Quest_Forward_x"].iloc[i], user_data["Oculus Quest_Forward_y"].iloc[i], user_data["Oculus Quest_Forward_z"].iloc[i])
        STIForwardX[i-2] = math.cos(-Ryaw) * forward_current.x - math.sin(-Ryaw) * forward_current.z
        STIForwardY[i-2] = forward_current.y
        STIForwardZ[i-2] = math.sin(-Ryaw) * forward_current.x + math.cos(-Ryaw) * forward_current.z

        STIfX_prev[i-2] = math.cos(-Ryaw) * F_prev.x - math.sin(-Ryaw) * F_prev.z
        STIfY_prev[i-2] = F_prev.y
        STIfZ_prev[i-2] = math.sin(-Ryaw) * F_prev.x + math.cos(-Ryaw) * F_prev.z

        # esto deben ser las etiquetas
        position3 = Vector3(user_data["Oculus Quest_DevicePosition_x"].iloc[i+1], user_data["Oculus Quest_DevicePosition_y"].iloc[i+1],user_data["Oculus Quest_DevicePosition_z"].iloc[i+1])
        F = position3 - position2

        #FEATURE: SHORT TERM INPUT DELTA POSITON (F).
        STIlabelfX[i-2] = math.cos(-Ryaw) * F.x - math.sin(-Ryaw) * F.z
        STIlabelfZ[i-2] = math.sin(-Ryaw) * F.x + math.cos(-Ryaw) * F.z

    featuresST = {}
    featuresST['STIpitch'] = STIpitch
    featuresST['STIyaw'] = STIyaw

    featuresST['STIForwardX'] = STIForwardX
    featuresST['STIForwardY'] = STIForwardY
    featuresST['STIForwardZ'] = STIForwardZ

    featuresST['STIvelocityX'] = STIvelocityX
    featuresST['STIvelocityY'] = STIvelocityY
    featuresST['STIvelocityZ'] = STIvelocityZ

    featuresST['STIfX_prev'] = STIfX_prev
    featuresST['STIfY_prev'] = STIfY_prev
    featuresST['STIfZ_prev'] = STIfZ_prev

    featuresST['STIlabelfX'] = STIlabelfX
    featuresST['STIlabelfZ'] = STIlabelfZ

    return pd.DataFrame(featuresST)
```
